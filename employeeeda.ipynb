```python
# employee_attrition_analysis.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("Employee-Attrition.csv")
print(f"âœ… Dataset Loaded: {df.shape}")

# --- Attrition distribution ---
attrition_counts = df['Attrition'].value_counts()
print("\nðŸ“Š Attrition Count:")
display(attrition_counts)

# Plot Attrition count
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Attrition', hue='Attrition', palette='pastel', legend=False)
plt.title("Attrition Distribution")
plt.show()

# ===================== #
# 1. Top Features Contributing to Attrition
# ===================== #
top_features = ["YearsAtCompany", "JobSatisfaction", "TotalWorkingYears", 
                "PerformanceRating", "StockOptionLevel", "JobInvolvement"]

for feature in top_features:
    if feature in df.columns:
        plt.figure(figsize=(6,4))
        sns.barplot(data=df, x="Attrition", y=feature, hue="Attrition",
                    palette="Set2", errorbar="sd", legend=False)
        plt.title(f"{feature} vs Attrition")
        plt.show()

# ===================== #
# 2. Outlier Analysis
# ===================== #
outlier_features = ["MonthlyIncome", "YearsAtCompany", "YearsSinceLastPromotion", "YearsInCurrentRole"]

for feature in outlier_features:
    if feature in df.columns:
        plt.figure(figsize=(6,4))
        sns.boxplot(x=df[feature], color="skyblue")
        plt.title(f"Boxplot of {feature}")
        plt.show()

# ===================== #
# 3. Various Factors Based on Attrition
# ===================== #
categorical_features = ["Department", "Gender", "MaritalStatus", "BusinessTravel", "EducationField", "JobRole"]

for feature in categorical_features:
    if feature in df.columns:
        plt.figure(figsize=(8,5))
        sns.countplot(data=df, y=feature, hue="Attrition", palette="Set2")
        plt.title(f"{feature} vs Attrition")
        plt.show()
```
